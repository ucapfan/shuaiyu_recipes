```python
import os
import pandas as pd

filePath = "custom-data"

f2 = open("test2.csv","a")
f2.write("specie,Record Number,Process,Theory or Experiment,data type,Projectile Element,Projectile Charge State,\
Target Molecule,Target Charge State,Initial Excited State B,Products,Original X Unit,Original Y Unit,Reference Number,Author(s),\
Title of Record,Journal Name,Volume and Issue No,Page Numbers,Date of Publication,Process Title\n")

for files in os.listdir(filePath):
    for file in os.listdir("{}/{}".format(filePath,files)):
        with open("{}/{}/{}".format(filePath,files,file)) as f:
            lines = f.readlines()       
            f2.write("{},".format(files))
            print(lines)
            for line in lines:
                print(line)
                f2.write(line.split(" =")[1].replace("\n","").replace(",","###"))
                f2.write(",")
            f2.write("\n")
f2.close()
                
            
```
