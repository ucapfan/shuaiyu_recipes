```python
import os
import pandas as pd

filePath = "custom-data"

f2 = open("test.csv","a")
f2.write("specie,Id Number,Process,Element A,Initial Charge Number A,Final Charge Number A,Element B,Initial Charge Number B,\
            Final Charge Number B,Theory or Experiment,X Axis Data,Y Axis Data,Y Axis Error Data,Reference Number,\
            Author(s),Title of Record,Journal Name,Volume and Issue No,Page Numbers,Date of Publication,Process Title\n")

for files in os.listdir(filePath):
    for file in os.listdir("{}/{}".format(filePath,files)):
        with open("{}/{}/{}".format(filePath,files,file)) as f:
            lines = f.readlines()       
            f2.write("{},".format(files))
            for line in lines:
                print(line)
                f2.write(line.split(" =")[1].replace("\n","").replace(",","###"))
                f2.write(",")
            f2.write("\n")
f2.close()
                
            
```
